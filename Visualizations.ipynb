```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/users/lakshmipriyanka/downloads/Linear Regression/BlackFriday.csv')
```

```python
df.fillna(value=0,inplace=True)
df["PC_2"] = df["PC_2"].astype(int)
df["PC_3"] = df["PC_3"].astype(int)
print('PC_2', df['PC_2'].unique())
print('PC_3', df['PC_3'].unique())
```

```python
sns.countplot(df['Gender'])
plt.xlabel('Gender')
plt.ylabel('Number of Buyers')
```

```python
sns.countplot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Number of Buyers')
```

```python
def sliceOfPie(column, explode, labels):
    ax1 = plt.subplot()
    ax1.pie(df[column].value_counts(), explode = explode, labels = labels, autopct = '%1.1f%%', textprops = {'color': "w", 'fontsize': 14}, shadow = True, startangle = 90)
    ax1.axis('equal')
    plt.tight_layout()
    plt.title('City Distribution')
    plt.legend()
```

```python
explode = (0.1, 0.1, 0.1)
labels = df['City_Category'].unique()
sliceOfPie('City_Category', explode, labels)
```
